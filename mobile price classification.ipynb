```python
X=Df1.drop(['price_range'],axis=1)
Y=Df1['price_range']
```

```python
sns.histplot(data=X['battery_power'],bins=15,kde=True)
```

```python
sns.histplot(data=X['int_memory'],bins=15,kde=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the features
X_Scaled = scaler.fit_transform(X)

# Convert back to DataFrame for better readability
X_scaled_df = pd.DataFrame(X_Scaled, columns=X.columns)

# Display the scaled features
print(X_scaled_df.head())
```

```python
cor=X_scaled_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(cor, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
X_scaled_df=X_scaled_df.drop(['px_height','px_width'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df,Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
